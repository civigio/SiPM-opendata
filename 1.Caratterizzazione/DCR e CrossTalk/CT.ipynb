```python
# DCR frequencies at different V_bias (same as DCR.ipynb)
frequency_Vb_54 = np.mean(Vb_54[0:7])
err_Vb_54 = np.std(Vb_54[0:7])/np.sqrt(len(Vb_54[0:7]))

frequency_Vb_54_5 = np.mean(Vb_54_5[0:10])
err_Vb_54_5 = np.std(Vb_54_5[0:10])/np.sqrt(len(Vb_54_5[0:10]))

frequency_Vb_55 = np.mean(Vb_55[0:12])
err_Vb_55 = np.std(Vb_55[0:12])/np.sqrt(len(Vb_55[0:12]))

frequency_Vb_55_5 = np.mean(Vb_55_5[0:15])
err_Vb_55_5 = np.std(Vb_55_5[0:15])/np.sqrt(len(Vb_55_5[0:15]))

frequency_Vb_56 = np.mean(Vb_56[0:17])
err_Vb_56 = np.std(Vb_56[0:17])/np.sqrt(len(Vb_56[0:17]))

frequency_Vb_56_5 = np.mean(Vb_56_5[0:19])
err_Vb_56_5 = np.std(Vb_56_5[0:19])/np.sqrt(len(Vb_56_5[0:19]))

frequency_Vb_57 = np.mean(Vb_57[0:21])
err_Vb_57 = np.std(Vb_57[0:21])/np.sqrt(len(Vb_57[0:21]))


# crosstalk frequencies at different V_bias
frequency_Vb_54_CST = np.mean(Vb_54[10:16])
err_Vb_54_CST = np.std(Vb_54[10:16])/np.sqrt(len(Vb_54[10:16]))

frequency_Vb_54_5_CST = np.mean(Vb_54_5[13:20])
err_Vb_54_5_CST = np.std(Vb_54_5[13:20])/np.sqrt(len(Vb_54_5[13:20]))

frequency_Vb_55_CST = np.mean(Vb_55[15:23])
err_Vb_55_CST = np.std(Vb_55[15:23])/np.sqrt(len(Vb_55[15:23]))

frequency_Vb_55_5_CST = np.mean(Vb_55_5[18:26])
err_Vb_55_5_CST = np.std(Vb_55_5[18:26])/np.sqrt(len(Vb_55_5[18:26]))

frequency_Vb_56_CST = np.mean(Vb_56[21:29])
err_Vb_56_CST = np.std(Vb_56[21:29])/np.sqrt(len(Vb_56[21:29]))

frequency_Vb_56_5_CST = np.mean(Vb_56_5[23:34])
err_Vb_56_5_CST = np.std(Vb_56_5[23:34])/np.sqrt(len(Vb_56_5[23:34]))

frequency_Vb_57_CST = np.mean(Vb_57[25:38])
err_Vb_57_CST = np.std(Vb_57[25:38])/np.sqrt(len(Vb_57[25:38]))

print(f'''Frequencies at different V_bias: 
    \n54 V: {frequency_Vb_54} ± {err_Vb_54} 
    \n54.5 V: {frequency_Vb_54_5} ± {err_Vb_54_5} 
    \n55 V: {frequency_Vb_55} ± {err_Vb_55} 
    \n55.5 V: {frequency_Vb_55_5} ± {err_Vb_55_5} 
    \n56 V: {frequency_Vb_56} ± {err_Vb_56} 
    \n56.5 V: {frequency_Vb_56_5} ± {err_Vb_56_5} 
    \n57 V: {frequency_Vb_57} ± {err_Vb_57}''')

print('\n============================\n')

# now we calculate the crosstalk probability for each V_bias as the ratio between the crosstalk frequency and the DCR frequency 
# NOTE: we assume that afterpulses do not fall in the gate. This is a good assumption since the afterpulses rarely falls in the gate.

def error_propagation(f_DCR, f_CST, err_f_DCR, err_f_CST):
    return np.sqrt((err_f_CST/f_DCR)**2 + (f_CST*err_f_DCR/f_DCR**2)**2)

probabilities = []
err_probabilities = []


p_54 = frequency_Vb_54_CST/frequency_Vb_54
err_p_54 = error_propagation(frequency_Vb_54, frequency_Vb_54_CST, err_Vb_54, err_Vb_54_CST)
probabilities.append(p_54)
err_probabilities.append(err_p_54)

p_54_5 = frequency_Vb_54_5_CST/frequency_Vb_54_5
err_p_54_5 = error_propagation(frequency_Vb_54_5, frequency_Vb_54_5_CST, err_Vb_54_5, err_Vb_54_5_CST)
probabilities.append(p_54_5)
err_probabilities.append(err_p_54_5)

p_55 = frequency_Vb_55_CST/frequency_Vb_55
err_p_55 = error_propagation(frequency_Vb_55, frequency_Vb_55_CST, err_Vb_55, err_Vb_55_CST)
probabilities.append(p_55)
err_probabilities.append(err_p_55)

p_55_5 = frequency_Vb_55_5_CST/frequency_Vb_55_5
err_p_55_5 = error_propagation(frequency_Vb_55_5, frequency_Vb_55_5_CST, err_Vb_55_5, err_Vb_55_5_CST)
probabilities.append(p_55_5)
err_probabilities.append(err_p_55_5)

p_56 = frequency_Vb_56_CST/frequency_Vb_56
err_p_56 = error_propagation(frequency_Vb_56, frequency_Vb_56_CST, err_Vb_56, err_Vb_56_CST)
probabilities.append(p_56) 
err_probabilities.append(err_p_56)

p_56_5 = frequency_Vb_56_5_CST/frequency_Vb_56_5
err_p_56_5 = error_propagation(frequency_Vb_56_5, frequency_Vb_56_5_CST, err_Vb_56_5, err_Vb_56_5_CST)
probabilities.append(p_56_5)
err_probabilities.append(err_p_56_5)

p_57 = frequency_Vb_57_CST/frequency_Vb_57
err_p_57 = error_propagation(frequency_Vb_57, frequency_Vb_57_CST, err_Vb_57, err_Vb_57_CST)
probabilities.append(p_57)
err_probabilities.append(err_p_57)

print(f'Probability of CST at V_bias = 54 V: {p_54:.3f} ± {err_p_54:.3f}')
print(f'Probability of CST at V_bias = 54.5 V: {p_54_5:.3f} ± {err_p_54_5:.3f}')
print(f'Probability of CST at V_bias = 55 V: {p_55:.3f} ± {err_p_55:.3f}')
print(f'Probability of CST at V_bias = 55.5 V: {p_55_5:.3f} ± {err_p_55_5:.3f}')
print(f'Probability of CST at V_bias = 56 V: {p_56:.3f} ± {err_p_56:.3f}')
print(f'Probability of CST at V_bias = 56.5 V: {p_56_5:.3f} ± {err_p_56_5:.3f}')
print(f'Probability of CST at V_bias = 57 V: {p_57:.3f} ± {err_p_57:.3f}')
```

occhio che dovrò usare overvoltage \Delta V = V_bias - V_breakdown; come crosstalk probability invece devo fare frequency_2nd_plateau / frequency_1st_plateau

```python
import numpy as np
import jax
from iminuit import Minuit
from jax import grad
from scipy.stats import chi2

V_breakdown = 51.475
overvoltage = np.array([54, 54.5, 55, 55.5, 56, 56.5, 57]) - V_breakdown
probabilities = np.array(probabilities)
err_probabilities = np.array(err_probabilities)
overvoltage_err = np.ones(len(overvoltage)) * 0.485 # lo stesso di V_breakdown trovato in DCR.ipynb

jax.config.update("jax_enable_x64", True)

def model(x, k):
    return 1 - jax.numpy.exp(-k*x**2)

f_prime = jax.jit(grad(model))

@jax.jit
def cost(k):
    result = 0.0
    for xi, yi, y_erri, x_erri in zip(overvoltage, probabilities, err_probabilities, overvoltage_err):
        y_var = y_erri ** 2 + (f_prime(xi, k)*x_erri) ** 2
        result += (yi - model(xi, k)) ** 2 / y_var
    return result

cost.errordef = Minuit.LEAST_SQUARES

my_minuit = Minuit(cost,k=0.003)
my_minuit.migrad()
my_minuit.hesse()
display(my_minuit)

# Calcolo del numero di gradi di libertà (ndof)
ndof = len(probabilities) - 1

# Calcolo del p-value utilizzando la distribuzione chi-quadro
p_value = (1. - chi2.cdf(my_minuit.fval, df=ndof))

# Calcolo del chi-quadro ridotto
chi_reduced = my_minuit.fval / ndof

print(f"p-value: {p_value}")
print(f"chi-quadro ridotto: {chi_reduced}")

print(f'overvoltage = {overvoltage[0]} +/- {overvoltage_err[0]}')
print(f'overvoltage = {overvoltage[1]} +/- {overvoltage_err[1]}')
print(f'overvoltage = {overvoltage[2]} +/- {overvoltage_err[2]}')
print(f'overvoltage = {overvoltage[3]} +/- {overvoltage_err[3]}')
print(f'overvoltage = {overvoltage[4]} +/- {overvoltage_err[4]}')
print(f'overvoltage = {overvoltage[5]} +/- {overvoltage_err[5]}')
print(f'overvoltage = {overvoltage[6]} +/- {overvoltage_err[6]}')
```

```python
xaxis = np.linspace(0, 7, 10000)
plt.figure(figsize=(10,6))
plt.errorbar(overvoltage, probabilities, xerr=overvoltage_err, yerr=err_probabilities, fmt='.', ecolor='black', label='Data', color='black')
plt.plot(xaxis, model(xaxis, my_minuit.values['k']), label='Fit', color='blue')

fit_info = [f"$\\chi^2$/$n_\\mathrm{{dof}}$ = {my_minuit.fval:.1f} / {ndof:.0f} = {chi_reduced:.1f}",]
fit_info.append(f"P-value = {p_value:.3f}")
fit_info.append('$k = $ ({0:.2f} $\\pm$ {1:.2f})'.format(my_minuit.values['k']*10**3, my_minuit.errors['k']*10**3) + ' $ \cdot 10^{-3} $ $V^{-2}$')
fit_info.append('$CP(\Delta V) = 1 - e^{-k {\Delta V}^2}$')
plt.legend(title="\n".join(fit_info), frameon=True)
plt.title('Crosstalk probability vs Overvoltage')
plt.grid(True)
plt.xlabel('$\Delta V$ $[V]$')
plt.ylabel('Crosstalk probability CP')
```
