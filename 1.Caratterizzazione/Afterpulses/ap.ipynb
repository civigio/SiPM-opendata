```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
        
apdata = pd.read_csv('apdata.CSV', sep='\t')
apdata
```

```python
x = np.array(apdata['Channel'])
y_on = np.array(apdata['On'])
y_off = np.array(apdata['Off'])
```

```python
import plotly.express as px
fig = px.scatter(apdata, x='Channel', y='On')
fig.add_scatter(x=apdata['Channel'], y=apdata['Off'], mode='markers')
fig.update_layout(yaxis_type='log')
fig.show()
```

```python
def sfoltisci_dati(x, y, num_punti):
    """
    Riduce il numero di punti nei dati x e y mantenendo la distribuzione.
    
    Parameters:
    - x: array numpy dei valori x
    - y: array numpy dei valori y
    - num_punti: numero desiderato di punti finali
    
    Returns:
    - x_sfoltito, y_sfoltito: array numpy sfoltiti
    """
    if num_punti >= len(x):
        print("Il numero di punti richiesto Ã¨ maggiore o uguale ai punti originali. Nessuna modifica effettuata.")
        return x, y
    
    # Calcolo dell'indice per il campionamento
    indices = np.linspace(0, len(x) - 1, num=num_punti, dtype=int)
    
    # Campionamento dei dati
    x_sfoltito = x[indices]
    y_sfoltito = y[indices]
    
    return x_sfoltito, y_sfoltito
```

```python
s_x_on, s_y_on = sfoltisci_dati(x, y_on, int(len(x)/7))
s_x_off, s_y_off = sfoltisci_dati(x, y_off, int(len(x)/7))
```

```python
fig, ax = plt.subplots()
#dimensioni figura
fig.set_figwidth(15)
ax.errorbar(s_x_on, s_y_on, yerr=np.sqrt(s_y_on), fmt='-', label='Light ON')
ax.errorbar(s_x_off, s_y_off, yerr=np.sqrt(s_y_off), fmt='-', label='Light OFF')
#ax.errorbar(s_x_off, y=np.abs(s_y_on-s_y_off), yerr=np.sqrt(np.abs(s_y_on-s_y_off)), fmt='-', label='Difference')
ax.set_xlabel('Channel')
ax.set_ylabel('Counts')
plt.title('Afterpluses - internal trigger')
ax.set_yscale('log')
ax.legend()
ax.grid()
plt.show()
fig.savefig('ap-total-log.png', dpi=600)
```
